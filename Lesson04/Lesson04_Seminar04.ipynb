## Задача 3.

### Веса изделий следуют нормальному распределению с дисперсией 9 и средним арифметическим 50. Веса ниже 45 г считаются браком. Какая доля изделий бракованная?

$D(X) = 9 \\
 \sigma = 3 \\
 M(X) = 50$

$z_i = \cfrac{x_i - \bar{x}}{\sigma_x}$, \
$z_i$ - преобразованное значение исходного распределения к Z-стандарту \
$x_i$ - исходное наблюдение \
$\bar{x}$ - выборочное среднее

$z_i = \cfrac{x_i - \bar{x}}{\sigma_x} = \cfrac{45 - 50}{3} = -1.67 \\
 P(X < -1.67) = 0.0475$

## Задача 4.

### Веса изделий следуют нормальному распределению с дисперсией 9 и средним арифметическим 50. Найдите долю изделий, с весом:

### 1) выше 56 г

$z_i = \cfrac{x_i - \bar{x}}{\sigma_x} = \cfrac{56 - 50}{3} = 2 \\
P(X > 2) = 0.0228$

### 2) меньше 55 г

$z_i = \cfrac{x_i - \bar{x}}{\sigma_x} = \cfrac{55 - 50}{3} = 1.67 \\
P(X < 1.67) = 0.953$

### 3) не больше 50 г

$z_i = \cfrac{x_i - \bar{x}}{\sigma_x} = \cfrac{50 - 50}{3} = 0 \\
P(X < 0) = 0.5$

### 4) не больше 55 г, но не меньше 45 г

$z_1 = \cfrac{x_1 - \bar{x}}{\sigma_x} = \cfrac{45 - 50}{3} = -1.67 \\
 z_2 = \cfrac{x_2 - \bar{x}}{\sigma_x} = \cfrac{55 - 50}{3} = 1.67 \\
 P(45 < X < 55) = 0.953 - 0.0475 = 0.9055‬$

### 5) не меньше 48 г

$z_i = \cfrac{x_i - \bar{x}}{\sigma_x} = \cfrac{48 - 50}{3} = -0.67 \\
 P(X > -0.67) = 0.749$

### 6) меньше 54 г

$z_i = \cfrac{x_i - \bar{x}}{\sigma_x} = \cfrac{54 - 50}{3} = 1.33 \\
 P(X < 1.33) = 0.908$

### 7) больше 46 г

$z_i = \cfrac{x_i - \bar{x}}{\sigma_x} = \cfrac{46 - 50}{3} = -1.33 \\
 P(X > -1.33) = 0.908$

### 8) больше 52 г

$z_i = \cfrac{x_i - \bar{x}}{\sigma_x} = \cfrac{52 - 50}{3} = 0.67 \\
 P(X > 0.67) = 0.251$

### 9) от 44 до 54 г

$z_1 = \cfrac{x_1 - \bar{x}}{\sigma_x} = \cfrac{44 - 50}{3} = -2 \\
 z_2 = \cfrac{x_2 - \bar{x}}{\sigma_x} = \cfrac{54 - 50}{3} = 1.33 \\
 P(44 < X < 54) = 0.882‬$
