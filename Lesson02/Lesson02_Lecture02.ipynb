# Урок 2. Дискретные распределения вероятностей

**Случайная величина** \
— численная характеристика случайного эксперимента. Её значение заранее неизвестно и зависит от свершившегося исхода.

***
## Дискретная случайная величина

**Дискретная** (прерывная) случайная величина \
– принимает отдельно взятые, изолированные значения. Количество этих значений конечно либо бесконечно, но счётно.

***
### Математическое ожидание

**Математическое ожидание** – это среднее значение случайной величины, при числе испытаний стремящихся к бесконечности \
$M(X) = n \cdot p$

***
### Дисперсия

**Дисперсия** характеризует степень рассеянности значений случайной величины относительно ее математического ожидания \
$D(X) = n \cdot p \cdot q$

***
### Среднее квадратичное отклонение

**Среднее квадратичное отклонение** показывает, насколько далеко наблюдения могут быть "разбросаны" относительно их среднего значения $\mu$ \
$σ = \sqrt{D(X)}$

***
## Распределения дискретной случайной величины

***
### Биномиальное распределение. Формула Бернулли

Случайная дискретная величина Х имеет **биномиальное распределение** с переменными $k$, $n$, $p$, если вероятность ее распределения задается следующим уравнением: \
$P_n(X=k) = C_n^k \cdot p^k \cdot q^{n-k}$, где \
$k$ - число наступления события (дискретная величина из отрезка $[0,n]$) \
$n$ - число испытаний \
$p$ - вероятность наступления события $A$ в независимых испытаниях \
$q = 1 - p$

#### Наивероятнейшее число появления события

Число $k_0$, которому при заданном $n$ соответствует максимальная биноминальная вероятность $P_n(k_0)$, называется **наивероятнейшим** числом появления события $A$. При заданных $n$ и $p$ это число определяется неравенствами: \
$n \cdot p-q \leq k_0 \leq n \cdot p + p$

Если $n \cdot p + p \in R$, то $k_0 = $ целой части.\
Если $n \cdot p + p \in Z$, то $k_0$ принимает два значения: $n \cdot p-q$ и $n \cdot p + p$.

***
### Распределение Пуассона

Если проводится большое количество испытаний $n$ и при этом вероятность $p$ появления события $A$ в отдельном испытании мала ($<0,1$), то вероятность того, что событие произойдет $m$ раз в $n$ испытаний рассчитывают по формуле: \
$P_m \approx \cfrac{λ^m}{m!} \cdot e^{-λ}$, где $λ$ - среднее количество наступления события за определенную единицу измерений.
